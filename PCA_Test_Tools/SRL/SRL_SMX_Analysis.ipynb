### EXTRACT VIPER ID, RANGE GROUND TRUTH, TARGET SIZE GROUND TRUTH AND TARGET TYPE

```python
def directory_data_extractor(directory):
 indx=0
 for i in directory:
    indx=indx+1
    if directory[indx:indx+3]=='VPR' or directory[indx:indx+3]=='vpr':
        viper_id = directory[indx+4:indx+6]
        viper_id = 'VPR0'+viper_id
        print('Viper ID is:',viper_id)
        break

 indx=0
 for i in directory:
    indx=indx+1
    if directory[indx:indx+2]=='GT' or directory[indx:indx+2]=='gt':
        if float(directory[indx+2])<10 and directory[indx+4]!='.':
         ground_truth_str = directory[indx+2:indx+6]
         ground_truth = float(ground_truth_str)
         print('Ground Truth Range is:',ground_truth)
         break
        else:
         ground_truth_str = directory[indx+2:indx+7]
         ground_truth = float(ground_truth_str)
         print('Ground Truth Range is:',ground_truth)
         break

 indx=0
 for i in directory:
    indx=indx+1
    if directory[indx:indx+6]=='Target' or directory[indx:indx+6]=='target':
        
        if directory[indx+6]=='0':
            size_a=float(directory[indx+7:indx+10])
            size_b=float(directory[indx+12:indx+15])
            gt_target_size_area= size_a*size_b
            target_type = 'Lambertian'
            print('Target Size Ground Truth Area:',gt_target_size_area)
            print("Target is a type:",target_type)
            break
        else:
         gt_target_size_area = float(directory[indx+6])*float(directory[indx+8])
         target_type = 'Lambertian'
         print('Target Size Ground Truth Area:',gt_target_size_area)
         print("Target is a type:",target_type)
         break
    elif directory[indx:indx+6] == 'Stop24' or directory[indx:indx+4] == 'stop24':
        gt_target_size_area = 0.31397
        target_type = '24" Stop'
        print('Target Size Ground Truth Area:',gt_target_size_area)
        print(target_type)
        break
    elif directory[indx:indx+6] == 'Stop30' or directory[indx:indx+4] == 'stop30':
        gt_target_size_area = 0.49443
        target_type = '30" Stop'
        print('Target Size Ground Truth Area:',gt_target_size_area)
        print(target_type)
        break
    elif directory[indx:indx+12] == 'LostCargo_5p' or directory[indx:indx+12] == 'lostcargo_5p':
       target_type = 'LC 5p'
       gt_target_size_area = 0.31
       print('Target Size Ground Truth Width:',gt_target_size_area)
       print('Target Type:',target_type)
    elif directory[indx:indx+13] == 'LostCargo_10p' or directory[indx:indx+13] == 'lostcargo_10p':
       target_type = 'LC 10p'
       gt_target_size_area = 0.5
       print('Target Size Ground Truth Width:',gt_target_size_area)
       print('Target Type:',target_type)
    elif directory[indx:indx+13] == 'KneelingChild' or directory[indx:indx+13] == 'kneelingchild':
       target_type = 'Child Dummy'
       gt_target_size_area = 0.720
       print('Target Size Ground Truth Heightg:',gt_target_size_area)
       print('Target Type:',target_type)
    elif directory[indx:indx+11] == 'SmallObject' or directory[indx:indx+13] == 'smallobject':
       target_type = 'Small Object'
       gt_target_size_area = 0.1
       print('Target Size Ground Truth Width:',gt_target_size_area)
       print('Target Type:',target_type)
    elif directory[indx:indx+4] == 'Curb' or directory[indx:indx+13] == 'curb':
       target_type = 'Curb'
       gt_target_size_area = 0.18
       print('Target Size Ground Truth Height:',gt_target_size_area)
       print('Target Type:',target_type)

 indx=0
 for i in directory:
    indx=indx+1
    if directory[indx:indx+3]=='PFA' or directory[indx:indx+6]=='pfa':
        target_type = 'Open Sky'
        print(target_type)

 indx=0
 for i in directory:
    indx=indx+1
    if directory[indx:indx+4]=='2024' or directory[indx:indx+4]=='2023'or directory[indx:indx+4]=='2025':
        test_date = directory[indx+5:indx+7]+"/"+directory[indx+8:indx+10]+"/"+directory[indx+2:indx+4]
        print("Test Date:",test_date)
        
 if target_type == 'Open Sky':
  return viper_id,target_type,test_date
 else:
  return viper_id, ground_truth, gt_target_size_area, target_type, test_date
```

```python
# Plotting
plt.figure(figsize=(10, 6))
# plt.bar(df_range_results_copy['test_case_target'], df_range_results_copy['90% PD Points'], color='skyblue')


bars = plt.bar(df_range_results_copy['test_case_target'], df_range_results_copy['90% PD Points'], color=df_range_results_copy['Color'])

# Adding labels and title
plt.xlabel('Distance and Target Type')
plt.ylabel('Number of Points')
plt.title('90% PD Point Count on Targets per Distance SW 1.4.1')
plt.xticks(rotation=45, ha='right')  # Rotate x labels for better readability

# Displaying the amount of points on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom')


# Show the plot
plt.tight_layout()
plt.show()
```
